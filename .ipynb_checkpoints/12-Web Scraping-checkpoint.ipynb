### <span style = 'color:magenta' > Getting Text from a Website </span>

```python
from bs4 import BeautifulSoup
import requests
url = 'https://webscraper.io/test-sites/e-commerce/allinone/phones'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
all_phone_prices = soup.find_all('h4', {'class':'price float-end card-title pull-right'})
```

```python
all_phone_titles = soup.find_all('a', 'title')
```

```python
all_prices = [prices.text for prices in all_phone_prices]
print(all_prices)
```

```python
all_titles = [titles.text for titles in all_phone_titles]
print(all_titles)
```

```python
import pandas as pd
pd.DataFrame(all_prices, all_titles, columns = ['Prices'])
```
